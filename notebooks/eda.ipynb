```python


plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Survived"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=["Not Survived", "Survived"]
)
plt.title("Survival Percentage")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Pclass"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Passenger Class Distribution")
plt.ylabel("")
plt.show()
```

```python
fare = df.groupby("Pclass")["Fare"].mean()

plt.figure(figsize=(6,4))
plt.plot(fare.index, fare.values, marker="o")
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.show()
```

```python
age = df.groupby("Pclass")["Age"].mean()

plt.figure(figsize=(6,4))
plt.plot(age.index, age.values, marker="o")
plt.title("Average Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.show()
```

```python
survival = df.groupby("Sex")["Survived"].mean()

plt.figure(figsize=(6,4))
plt.bar(survival.index, survival.values)
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```
